```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_film = pd.read_csv('netflix_titles.csv')
df_film
```

```python
'''
    pertanyaan
    
1. genre film apa yang memiliki rata2 durasi paling lama
2. distribusi film berdasarkan durasi
3. 10 Negara mana yang paling banyak produksi film selain amerika dan india
4. distribusi film berdasarkan bulan
    spring(maret-mei), summer(juni-agustus), fall(sept-nov), winter(dec-februari)

'''
```

```python
df_film.isna().sum()
```

```python
df_film.drop('director', axis=1, inplace=True)
```

```python
df_film.dropna(inplace=True)
df_film
```

```python
df_film.drop('show_id', axis=1, inplace=True)
```

```python
df_film['type'].unique()
```

```python
df_film['rating'].unique()
```

```python
df_film['title'].nunique()
```

```python
df_type = df_film.groupby('type')
df_genre_new = df_type.get_group('Movie')[['title','listed_in', 'duration']]
df_genre_movie = pd.DataFrame(df_genre_new)
df_genre_movie['duration'] = df_genre_movie['duration'].str.split().str[0].astype('int')
df_genre_movie['listed_in'] = df_genre_movie['listed_in'].str.replace('&',',')
df_genre_movie['listed_in'] = df_genre_movie['listed_in'].str.replace(' ','')
df_genre_movie
```

```python
df_genre_movie['listed_in'] = [i.split(',', len(i)) for i in list(df_genre_movie['listed_in'])]
df_genre_movie
```

```python
df_genre_movie = df_genre_movie.reset_index(drop=True)
df_genre_movie
```

```python
df_genre_movie['duration baru'] = df_genre_movie['duration'].astype('str')
df_genre_movie['duration baru'] = [list(df_genre_movie['duration baru'])for x in df_genre_movie['listed_in']]
df_genre_movie
```

```python
list_dur = list(zip(df_genre_movie['listed_in'], df_genre_movie['duration']))
list_dur
```

```python
dur = []
for i, v in list_dur:
    dur_i = []
    for x in range(len(i)):
        dur_i.append(v)
    dur.append(dur_i)
df_genre_movie['duration_baru'] = dur
df_genre_movie.drop('duration baru', axis=1, inplace=True)
df_genre_movie
```

```python
def genre(series):
 return pd.Series([x for _list in series for x in _list])
new_df = genre(df_genre_movie['listed_in'])
type(new_df)
```

```python
def genre(series):
 return pd.Series([x for _list in series for x in _list])
new_df_dur = genre(df_genre_movie['duration_baru'])
type(new_df_dur)
```

```python
df_baru = pd.concat([new_df,new_df_dur ], axis=1) #pertanyaan buat pak ridho kenapa harus pakai concat
(df_baru)
```

```python
# df_baru.groupby(0)
mean = df_baru.groupby(0).mean()
df_mean = pd.DataFrame(mean)
df_mean
# df_duration = pd.DataFrame(df_baru['Genre'].count(), df_baru['Genre'].count().index)
```

### Menjawab pertanyaan nomor 1: rata2 durasi film berdasarkan genre

```python
df_mean = df_mean.sort_values(1,ascending=False) #mengurutkan dari yang paling besar
plt.figure(figsize=(16,7))
plt.bar(df_mean.index, df_mean[1])
plt.xticks(rotation=90)
plt.title('Rata2 Durasi Film Berdasarkan Genre')
plt.ylabel('Durasi (menit)')
plt.show
```

### menjawab pertanyaan nomor 2: distribusi film berdasarkan durasi

```python
plt.figure(figsize=(16,7))
df_baru[1].hist(bins=15)
plt.xticks(rotation='vertical')
plt.axvline(df_baru[1].mean(), color='red')
plt.axvline(df_baru[1].median(), color='orange')
plt.show()
```

```python
df_country_new = df_type.get_group('Movie')[['country']]
df_country_new
df_country_movie = pd.DataFrame(df_country_new)
df_country_movie['country'] = df_country_movie['country'].str.replace(' ','')
df_country_movie['country'] = [i.split(',', len(i)) for i in list(df_country_movie['country'])]
df_country_movie
```

```python
type(df_country_movie['country'][0][0])
```

```python
df_country_movie = df_country_movie.reset_index(drop=True)
```

```python
def country(series):
 return pd.Series([x for _list in series for x in _list])
new_df_coun = country(df_country_movie['country'])
new_df_coun
```

```python
df_baru_coun = pd.DataFrame([new_df_coun]).T
df_baru_coun[1] = 1
df_baru_coun
```

```python
sum_coun = df_baru_coun.groupby(0).count()
sum_coun.drop('', axis=0, inplace=True)
sum_coun.drop('UnitedStates', axis=0, inplace=True)
sum_coun.drop('India', axis=0, inplace=True)
```

```python
sum_coun
```

### Menjawab pertanyaan no 3: 10 negara terbanyak produksi film selain Amerika dan India

```python
sum_10 = sum_coun.sort_values(1,ascending=False).head(10)
plt.figure(figsize=(16,7))
plt.bar(sum_10.index, sum_10[1])
plt.xticks(rotation='vertical')
plt.title('10 Negara yang memproduksi film terbanyak selain Amerika dan India')
plt.show()
```
